```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()

```

```python
#Calculating the school count

school_count = len(school_data_complete["school_name"].unique())
school_count
```

```python
#Calculating the student count

student_count = len(school_data_complete["Student ID"].unique())
student_count
```

```python
#Calculating the total budget

unique_budget = school_data_complete["budget"].unique()
total_budget = sum(unique_budget)
total_budget
```

```python
#Calculating the average math score

average_mathscore = school_data_complete["math_score"].mean()
average_mathscore
```

```python
#Calculating the average reading score

average_readingscore = school_data_complete["reading_score"].mean()
average_readingscore
```

```python
#Calculating the passing math score (equal or greater than 70%)

passing_mathscore = (school_data_complete[school_data_complete["math_score"]>=70]["student_name"].count()/student_count)*100
passing_mathscore
```

```python
#Calculating the passing reading score

passing_readingscore = (school_data_complete[school_data_complete["reading_score"]>=70]["student_name"].count()/student_count)*100
passing_readingscore
```

```python
#Calculating the overall pass rate (both reading and math)

overall_passing = (school_data_complete[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)]["school_name"].count()/student_count)*100
overall_passing
```

# Creating Summary Dataframe

```python
district_summary = {"Total Schools": school_count,
                   "Total Students": '{:,}'.format(student_count),
                   "Total Budget": '${:,.2f}'.format(total_budget),
                   "Average Math Score": '{:,.2f}'.format(average_mathscore),
                   "Average Reading Score": '{:,.2f}'.format(average_readingscore),
                    "% Passing Math": '{:,.4f}'.format(passing_mathscore),
                    "% Passing Reading": '{:,.4f}'.format(passing_readingscore),
                     "% Overall Passing":[round(overall_passing)]
                   }
district_summary_df = pd.DataFrame(district_summary)
district_summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Display the school_level_summary_df for reference purposes

school_level_summary_df
```

```python
#Creating labels of the bins

cut_labels_4 = ['<$585', '$584-629', '$630-644', '$645-675k']

#Creating the bins that represent each of the spending ranges
cut_bins = [0, 585, 630, 645, 675]

#school_level_summary_df = school_level_summary

#Creating the df that contains the new series with the spending ranges
school_level_summary_df['Spending Ranges(Per Student)'] = pd.cut(per_studentbudget, bins=cut_bins, labels=cut_labels_4)


#Display the dataframe
school_level_summary_df

```

```python
#Groupby spending ranges for avg math, avg reading score, % passing math, % passing reading, and overall % passing rate

spending_range_math_score = school_level_summary_df.groupby(["Spending Ranges(Per Student)"]).mean()["Average Math Score"]
spending_range_reading_score = school_level_summary_df.groupby(["Spending Ranges(Per Student)"]).mean()["Average Reading Score"]
spending_range_passing_math = school_level_summary_df.groupby(["Spending Ranges(Per Student)"]).mean()["% Passing Math"]
spending_range_passing_reading = school_level_summary_df.groupby(["Spending Ranges(Per Student)"]).mean()["% Passing Reading"]
overall_range_passing_spending = school_level_summary_df.groupby(["Spending Ranges(Per Student)"]).mean()["% Overall Passing"]
```

```python
# Assemble into DataFrame
scores_byschool_spending = pd.DataFrame({"Average Math Score" : spending_range_math_score.round(2),
                                 "Average Reading Score": spending_range_reading_score.round(2),
                                 "% Passing Math": spending_range_passing_math.round(2),
                                 "% Passing Reading": spending_range_passing_reading.round(2),
                                 "% Overall Passing": overall_range_passing_spending.round(2)})




# Display results
scores_byschool_spending
```
